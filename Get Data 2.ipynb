## Webscrape

```python
import requests
import pandas as pd

from pathlib import Path  
```

### Wikipedia

```python
filepath = Path('importdata/wikipedia.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_highest-grossing_Japanese_films'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/102.0.0.0 Safari/537.36'}
```

```python
result = requests.get(url, headers=headers).content
result
```

```python
tables = pd.read_html(result)
df = tables[0]
df
```

```python
df.to_csv(filepath)  
```

### eiren 

```python
filepath = Path('importdata/eiren.csv')  
filepath.parent.mkdir(parents=True, exist_ok=True)  
```

```python
url1 = 'http://www.eiren.org/boxoffice_e/' 
url2 = '.html'
```

```python
url1 = 'http://www.eiren.org/boxoffice_e/' 
url2 = '.html'
final_df = pd.DataFrame(columns = [0,1,2,3,4])

years = range(2004,2022)
for year in years:
    url = url1 + (str(year)) + url2
    
    result = requests.get(url, headers=headers).content
    tables = pd.read_html(result)

    final_df = final_df.append(tables[1])
```

```python
final_df.to_csv(filepath)  
```
